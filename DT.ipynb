```python
import pandas as pd 
df = pd.read_csv("50_Startups.csv")
df = pd.get_dummies(df,dtype=int,drop_first=True)
independant = df[["R&D Spend","Administration","Marketing Spend","State_Florida","State_New York"]]
dependant = df[["Profit"]]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independant,dependant,test_size=0.30,random_state=0)
 
x_train
```

```python
from sklearn.tree import DecisionTreeRegressor
DT = DecisionTreeRegressor(criterion= "absolute_error",random_state=1000)
DT
```

```python
DT.fit(x_train,y_train)
```

```python
y_predict = DT.predict(x_test)
from sklearn.metrics import r2_score
r2_score=r2_score(y_test,y_predict)
r2_score
```

```python
import pickle
filename = "finalized_model_MLR.sav"
pickle.dump(DT,open(filename,'wb'))
Loaded_model=pickle.load(open("finalized_model_MLR.sav",'rb'))
years=int(input("enter:"))
```
